```python
import matplotlib.pyplot as plt
import numpy as np

# New data
classes = [
    "NR-AR", "NR-AR-LBD", "NR-AhR", "NR-Aromatase", "NR-ER",
    "NR-ER-LBD", "NR-PPAR-gamma", "SR-ARE", "SR-ATAD5",
    "SR-HSE", "SR-MMP", "SR-p53"
]

roc_auc_scores = [
    0.6367, 0.7136, 0.7318, 0.8557, 0.6440,
    0.8170, 0.7757, 0.7306, 0.7948, 0.7220,
    0.7479, 0.8517
]

# Calculate Mean AUC
mean_auc = np.mean(roc_auc_scores)

# Plotting the bar chart
plt.figure(figsize=(10, 6))

# Remove label for bars to not appear in the legend
bars = plt.barh(classes, roc_auc_scores, color='skyblue')

# Add horizontal line for mean AUC
mean_line = plt.axvline(mean_auc, color='red', linestyle='--', label=f'Mean AUC = {mean_auc:.4f}')

# Add labels for each bar
for bar in bars:
    plt.text(bar.get_width() + 0.01, bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():.4f}', va='center', ha='left', fontsize=10)

# Labels and title
plt.xlabel('ROC-AUC Score')
plt.title('ROC-AUC Comparison for LSTM')
plt.xlim(0, 1)  # AUC score is between 0 and 1


# Display only the mean AUC in the legend, positioned at the bottom-right corner
plt.legend(loc='lower right')

# Show the plot
plt.show()
```
